1.   Genera 10000 punts que segueixin una distribució uniforme entre 0 i 10 i fes un histograma normalitzat.

2.   Genera 10000 punts que segueixin una distribució gaussiana centrada a 5 amb una $\sigma$ de 3 i fes un histograma normalitzat.

Pots fer servir les llibreries de scipy o de numpy:

```
import numpy as np
from scipy import stats
stats.norm.rvs(loc=5,scale=3,size=N)
stats.uniform.rvs(loc=0, scale=10, size=N)

```



````python
# El teu codiimport numpy as np  # Importa la librería para cálculos numéricos (alias 'np'). Se usa para crear arrays de datos.
import matplotlib.pyplot as plt  # Importa la herramienta para hacer gráficas (alias 'plt'). Su función .hist hace el dibujo.
from scipy import stats  # De la librería Scipy (ciencia), importa el módulo 'stats' que tiene las fórmulas de estadística.

N = 10000  # Definimos una variable con el número total de puntos. A más puntos, más suave se verá el dibujo.

# --- PARTE 1: DISTRIBUCIÓN UNIFORME ---

# np.random.uniform genera números donde todos tienen la misma probabilidad (entre low y high).
d_unif = np.random.uniform(low=0, high=10, size=N)

# stats.uniform.rvs hace lo mismo pero desde Scipy.
# 'loc' es el inicio (0) y 'scale' es el ancho del intervalo (0 + 10 = hasta el 10).
d_unif2 = stats.uniform.rvs(loc=0, scale=10, size=N)



# plt.hist crea el histograma:
# bins=25: Divide el rango 0-10 en 25 barritas horizontales.
# density=True: Normaliza (la suma de las áreas de las barras será 1). Esto crea la "densidad".
# alpha=0.5: Hace que el color sea un poco transparente (50%).
# label: El nombre que aparecerá luego en la leyenda de la gráfica.
plt.hist(d_unif2, bins=25, density=True, alpha=0.5, label="unif. scipy")

plt.xlabel("x")  # Pone el título al eje horizontal (las unidades de tus datos).
plt.ylabel("normalized counts")  # Pone título al eje vertical (la frecuencia normalizada).
plt.legend(loc='best')  # Muestra el recuadro con los nombres (labels). 'best' deja que Python busque el hueco vacío.
plt.show()  # Comando final para "imprimir" y mostrar la ventana con el gráfico en pantalla.


# --- PARTE 2: DISTRIBUCIÓN NORMAL (CAMPANA DE GAUSS) ---

# stats.norm.rvs genera la famosa campana.
# loc=5: Es la media (el centro de la campana).
# scale=3: Es la desviación típica (sigma). Si es grande, la campana es ancha; si es pequeña, es estrecha.
d_norm = stats.norm.rvs(loc=5, scale=3, size=N)



[Image of normal distribution bell curve with mean and standard deviation]


# Dibujamos el segundo histograma:
plt.hist(d_norm, bins=25, density=True, label="norm. scipy")
plt.xlabel("x")  # Etiqueta eje X.
plt.ylabel("normalized counts")  # Etiqueta eje Y.
plt.legend(loc='best')  # Muestra la leyenda arriba a la derecha (por defecto).
plt.show()  # Muestra la campana de Gauss generada.
````
